```python
# read web table into pandas DF
import pandas as pd

linkToFile='https://github.com/CienciaDeDatosEspacial/code_and_data/raw/main/data/carbonEmi_downloaded.csv'
carbon=pd.read_csv(linkToFile)
```

```python
#Elimino slug y ranking usando "drop"
byeColumns=['slug','ranking']

#El resultado
carbon.drop(columns=byeColumns,inplace=True)
#Ahora
carbon
```

```python
carbon.rename(columns = {'date_of_information':'carbon_date'}, inplace = True)
```

```python
#A pesar que no se ven los espacios a primera vista en los nombres de las columnas, los eliminamos por si acaso
carbon.columns = carbon.columns.str.replace(' ', '')
carbon
```

```python
#Solo dejar los números en la columna "carbon_date"
carbon['carbon_date']=carbon.carbon_date.str.extract(pat=r'(\d+)',expand=True)
```
